# Problem Statement

# Tenant Screening Application

    - Tenant screening is the process Property Owners use to evaluate tenants before renting a property to them.
    - The goal is to assess the applicants and also to ensure that they will be responsible tenants, who can pay rent on 
      time and take care of their property. 

# Tenant Screening Key Points
   **I am going to Collect basic information,like**
   
    - Name 
    - Contact Details 
    - Employment History 
    - Income 
    - Rental History.
    
   **Bank Data also Because I need to consider Whether he/she is Eligible to pay my rent on time or not**
   
    - Credit card scores, payment history, outstanding debts, and public financial records.
    - Background Check:
    - Criminal records or legal issues.

   **Employment and Income Verification:**
   
    - Based on his/her income ensuring they can cover rent.
    
   **References Check:**
   
    - Gathers insights from personal and professional Recommendation about the applicantâ€™s character and reliability.
      

# Benefits of Tenant Screening
    - Reduced Risk of Rent Delinquency
    - Lower Property Damage Risk
    - Increased Safety
    - Legal Protection

# Common Screening Criteria
    - Credit Score: Indicator of financial responsibility.
    - Income-to-Rent Ratio: Ideally, income should be at least three times the rent.
    - Rental History: Positive past landlord references and on-time payments.
    - Criminal Record: Serious or recent offenses can be a red flag.
    - Employment Stability: Consistent employment or steady income.

# Data Collection

**Data Description**

    - Name: Name of the applicant.
    - Contact Details: Contact number of the applicant.
    - Employment History: Duration of current employment in years.
    - Income: Annual income of the applicant.
    - Rental History: Duration of previous rental history in years.
    - Credit Score: Credit score of the applicant.
    - Payment History: Indicates whether the payment history is on-time or has late payments.
    - Outstanding Debts: Total outstanding debts in dollars.
    - Public Financial Records: Any public financial records.
    - Criminal Records: Indicates any criminal records (None, Minor, Serious).
    - Legal Issues: Any legal issues the applicant might have.
    - Employment Verification: Indicates if the employment information is verified.
    - Income Verification: Indicates if the income information is verified.
    - Personal References: Feedback from personal references (Positive, Neutral, Negative).
    - Professional References: Feedback from professional references (Positive, Neutral, Negative).
    - Good Tenant: Target variable indicating whether the tenant is considered good (1) or not (0).

```python
# Creating a Pipeline to automate the comparision of all models to get the best Model
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, make_scorer
import time
import sys
import pickle
from tqdm import tqdm
import time
import os
import pickle
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, make_scorer
```

```python
mlflow.set_experiment("TenantApplication")
```

```python
# Creating pipelines for different classifiers
pipelines = {
    'knn': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', KNeighborsClassifier())
    ]),
    'svm': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', SVC())
    ]),
    'logistic_regression': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', LogisticRegression())
    ]),
    'random_forest': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', RandomForestClassifier())
    ]),
    'decision_tree': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', DecisionTreeClassifier())
    ]),

}
```

```python
# Hyperparameter grids for GridSearchCV
parameters = {
    'knn': {
        'classifier__n_neighbors': [3, 5, 7, 9],
        'classifier__p': [1, 2]  # Manhattan or Euclidean distance
    },
    'svm': {
        'classifier__C': [0.1, 1, 10],
        'classifier__kernel': ['linear', 'rbf'],
        'classifier__gamma': [0.1, 1, 10]
    },
    'logistic_regression': {
        'classifier__C': [0.1, 1, 10],
        'classifier__solver': ['liblinear', 'saga']
    },
    'random_forest': {
        'classifier__n_estimators': [50, 100, 200],
        'classifier__max_depth': [None, 10, 20, 30]
    },
    'decision_tree': {
        'classifier__max_depth': [None, 10, 20, 30],
        'classifier__min_samples_split': [2, 5, 10]
    },

}
```

```python
scoring = make_scorer(accuracy_score)

for model_name, pipeline in pipelines.items():
    grid_search = GridSearchCV(pipeline, parameters[model_name], cv=5, verbose=1, n_jobs=-1, scoring=scoring)
    mlflow.sklearn.autolog(max_tuning_runs=None)
    with mlflow.start_run() as run:
        start_time = time.time()
        grid_search.fit(X_train_scaled, y_train)
        end_time = time.time()

        best_model = grid_search.best_estimator_
        train_score = grid_search.score(X_train_scaled, y_train)
        test_score = grid_search.score(X_test_scaled, y_test)
        fit_time = end_time - start_time
        model_size = sys.getsizeof(pickle.dumps(best_model))

        print("********** {} **********".format(model_name.upper()))
        print("Best Parameters: {}".format(grid_search.best_params_))
        print("Training Accuracy Score: {:.2f}%".format(train_score * 100))
        print("Test Accuracy Score: {:.2f}%".format(test_score * 100))
        print("Wall Time: {:.2f} seconds".format(fit_time))
        print("Model Size: {} Bytes".format(model_size))

        mlflow.log_param("Best Parameters", grid_search.best_params_)
        mlflow.log_metric("Training Accuracy Score", train_score)
        mlflow.log_metric("Test Accuracy Score", test_score)
        mlflow.log_metric("Fit Time", fit_time)
        mlflow.log_metric("Model Size", model_size)
```

```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np
```

```python
# Define the input fields
employment_history = widgets.Text(description='Employment History:', value='5 years')
income = widgets.IntText(description='Income:', value=60000)
rental_history = widgets.Text(description='Rental History:', value='2 years')
credit_score = widgets.IntText(description='Credit Score:', value=700)
payment_history = widgets.Text(description='Payment History:', value='On-time')
outstanding_debts = widgets.IntText(description='Outstanding Debts:', value=5000)
criminal_records = widgets.Text(description='Criminal Records:', value='good')
legal_issues = widgets.Text(description='Legal Issues:', value='no records')
employment_verification = widgets.Text(description='Employment Verification:', value='Verified')
income_verification = widgets.Text(description='Income Verification:', value='Verified')
personal_references = widgets.Text(description='Personal References:', value='Positive')
professional_references = widgets.Text(description='Professional References:', value='Positive')

display(employment_history, income, rental_history, credit_score, payment_history,
        outstanding_debts, criminal_records, legal_issues, employment_verification,
        income_verification, personal_references, professional_references)


#***************************Processing the Input Because the model is trained on this preprocessing steps**********************#
def preprocess_input():
    input_data = {
        'Employment History': [employment_history.value],
        'Income': [income.value],
        'Rental History': [rental_history.value],
        'Credit Score': [credit_score.value],
        'Payment History': [payment_history.value],
        'Outstanding Debts': [outstanding_debts.value],
        'Criminal Records': [criminal_records.value],
        'Legal Issues': [legal_issues.value],
        'Employment Verification': [employment_verification.value],
        'Income Verification': [income_verification.value],
        'Personal References': [personal_references.value],
        'Professional References': [professional_references.value],
    }
    
    input_df = pd.DataFrame(input_data)
    
    for column in categorical_columns:
        le = label_encoders[column]
        input_df[column] = le.transform(input_df[column])
    
    input_scaled = scaler.transform(input_df)
    
    return input_scaled

def predict_tenant():
    input_scaled = preprocess_input()
    prediction = model.predict(input_scaled)
    result = "Approved" if prediction[0] == 1 else "Rejected"
    print(f'Tenant Screening Result: {result}')

submit_button = widgets.Button(description='Submit')
submit_button.on_click(lambda x: predict_tenant())
display(submit_button)

```
